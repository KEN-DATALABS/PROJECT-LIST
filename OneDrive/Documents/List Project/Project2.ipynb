```python
annual_sales= [f'Sales_month_{i}'for i in range(1,13)]
df_drop= df.drop(annual_sales,axis=1)
df_drop.head()
```

```python
# Define the monthly sales columns
monthly_cols = [f'Sales_month_{i}' for i in range(1, 13)]

# Melt the DataFrame
df_long = df.melt(
    id_vars=['Product_id', 'Category'],  # Columns to keep as identifiers
    value_vars=monthly_cols,             # Columns to unpivot
    var_name='month',                    # Name for the new month column
    value_name='sales'                   # Name for the new sales column
)

# Clean up the month column to show just numbers
df_long['month'] = df_long['month'].str.replace('Sales_month_', '').astype(int)

print(df_long.head(10))
```
