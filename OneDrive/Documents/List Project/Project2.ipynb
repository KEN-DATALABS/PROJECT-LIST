```python
data = {
    'customer_id': [101, 102, 103, 104, 101, None],
    'customer name': ['Alice Smith', 'bob jones', 'Carol Wilson', 'DAVID BROWN', 'alice smith', 'Eva Davis'],
    'order_value': [150.0, None, 200.0, 180.5, 150.0, None],
    'city': ['New York', 'london', 'Paris', 'NEW YORK', 'New York', 'berlin']
}
df_cus = pd.DataFrame(data)
df_cus.head(10)
```

```python
df_cus.drop_duplicates(subset= ['customer name'],keep ='first')
```

```python
df_cus=df_cus.fillna({'order_value':df_cus['order_value'].mean()})
df_cus
```

```python
df_cus['customer_id']=df_cus['customer_id'].fillna(df_cus['customer_id'].mean())
df_cus
```

```python
df_cus
```

```python
df_cus['city']= df_cus['city'].str.title()
df_cus
```

```python
df_cus['customer name'] = df_cus['customer name'].str.title()
df_cus
```

```python
df_cus['email']= df_cus['customer name'].str.split().str[0] + '.' + df_cus['customer name'].str.split().str[1] + '@company.com'
df_cus
```

```python
def value(x):
    value = ''
    if x < 100:
        value = 'Small'
    elif 100 <= x < 200:
        value = 'Medium'
    elif x>=200:
        value = 'Large'
    return value 
df_cus['order_category'] = df_cus['order_value'].apply(lambda x: value(x))
df_cus
```

```python
df_cus['region']= df_cus['city'].apply(lambda x: 'Americas' if x == 'New York' else 'Europe')
df_cus
```

```python
df_cus['order_value'].dtypes
df_cus['order_value']= df_cus['order_value'].astype('int32')
df_cus
```

```python
# Sales data
sales = pd.DataFrame({
    'Region': ['North', 'South', 'North', 'South', 'East', 'West', 'East', 'West'],
    'Product': ['A', 'A', 'B', 'B', 'A', 'B', 'A', 'B'],
    'Salesperson': ['John', 'Jane', 'John', 'Jane', 'Mike', 'Sarah', 'Mike', 'Sarah'],
    'Revenue': [1000, 1200, 800, 900, 1100, 950, 1300, 850],
    'Units': [10, 12, 8, 9, 11, 10, 13, 9]
})
sales.head(10)
```

```python
sales.groupby('Region')['Revenue'].sum()
```

```python
sales.groupby('Product')['Revenue'].sum()
```

```python
sales.groupby('Salesperson').size()
```

```python
sales.groupby(['Region','Product'])['Revenue'].max().reset_index()
```

```python
sales.groupby(['Region','Product']).agg({
    'Revenue':'sum',
    'Units':'sum'})
```

```python
sales.groupby(['Region','Salesperson'])['Revenue'].mean().reset_index()
```

```python
result = sales.groupby('Product').agg({'Revenue':[lambda x: x.max()-x.min()]})
result.columns = ['_'.join(col).split() for col in result.columns]
result
```
