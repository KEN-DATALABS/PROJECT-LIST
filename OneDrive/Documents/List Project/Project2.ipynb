```python
data = {
    'customer_id': [101, 102, 103, 104, 101, None],
    'customer name': ['Alice Smith', 'bob jones', 'Carol Wilson', 'DAVID BROWN', 'alice smith', 'Eva Davis'],
    'order_value': [150.0, None, 200.0, 180.5, 150.0, None],
    'city': ['New York', 'london', 'Paris', 'NEW YORK', 'New York', 'berlin']
}
df_cus = pd.DataFrame(data)
df_cus.head(10)
```

```python
df_cus.drop_duplicates(subset= ['customer name'],keep ='first')
```

```python
df_cus
```

```python
df_cus['city']= df_cus['city'].str.title()
df_cus
```

```python
df_cus['customer name'] = df_cus['customer name'].str.title()
df_cus
```

```python
df_cus['email']= df_cus['customer name'].str.split().str[0] + '.' + df_cus['customer name'].str.split().str[1] + '@company.com'
df_cus
```

```python
def value(x):
    value = ''
    if x < 100:
        value = 'Small'
    elif 100 <= x < 200:
        value = 'Medium'
    elif x>=200:
        value = 'Large'
    return value 
df_cus['order_category'] = df_cus['order_value'].apply(lambda x: value(x))
df_cus
```

```python
df_cus['region']= df_cus['city'].apply(lambda x: 'Americas' if x == 'New York' else 'Europe')
df_cus
```

```python
wide_sales = pd.DataFrame({
    'Region': ['North', 'South'],
    'Product_A_2022': [1000, 1200],
    'Product_B_2022': [800, 900],
    'Product_A_2023': [1100, 1300],
    'Product_B_2023': [850, 950]
})
wide_sales 
```

```python
long =pd.melt(wide_sales,
        id_vars = ['Region'],
        value_vars = ['Product_A_2022','Product_B_2022','Product_A_2023','Product_B_2023'],
        var_name = 'Product',
        value_name = 'Sales')
long
```

```python
long['Year'] = long['Product'].str.split('_').str[2]
long
```

```python
long['Product'] = long['Product'].str.split('_').str[0]+'_'+long['Product'].str.split('_').str[1]
long
```

```python
unstack= pd.pivot_table(long,
                        values = 'Sales',
                        columns = 'Year',
                        index = ['Region','Product'],
                        aggfunc = 'sum')
unstack
```
