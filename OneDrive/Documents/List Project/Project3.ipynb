## **Task 5: Practice Project - "Student Performance Tracker"**

```python
class_t={
'Name':['ken','paul','bolu','chuka','bola','peter'],
'Math_Score':[12,None,67,32,18,None],
'Science_Score':[92,18,None,47,28,62],
'English_Score':[74,None,76,65,81,71]}
class_track= pd.DataFrame(class_t)
class_track
```

```python
class_track['Name']=class_track['Name'].str.capitalize()
class_track.head()
```

```python
class_track['Avg_Score']= class_track[['Math_Score','Science_Score','English_Score']].mean(axis=1)
class_track.head()
```

```python
class_track[class_track['Math_Score']>20]
```

```python
class_track['Total_Score']=class_track[['Math_Score','Science_Score','English_Score']].sum(axis=1)
class_track.sort_values(by='Total_Score',ascending= True)
```
