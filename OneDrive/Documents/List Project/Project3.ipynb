### **Task 5: Practice Project - "Student Performance Tracker"**

```python
class_t={
'Name':['ken','paul','bolu','chuka','bola','peter'],
'Math_Score':[12,None,67,32,18,None],
'Science_Score':[92,18,None,47,28,62],
'English_Score':[74,None,76,65,81,71]}
class_track= pd.DataFrame(class_t)
class_track
```

```python
class_track['Name']=class_track['Name'].str.capitalize()
class_track.head()
```

```python
class_track['Avg_Score']= class_track[['Math_Score','Science_Score','English_Score']].mean(axis=1)
class_track.head()
```

```python
class_track[class_track['Math_Score']>20]
```

```python
class_track['Total_Score']=class_track[['Math_Score','Science_Score','English_Score']].sum(axis=1)
class_track.sort_values(by='Total_Score',ascending= True)
```

# **Phase 2 Tasks**

### **Task 1: Data Cleaning**

```python
data = {
    'customer_id': [101, 102, 103, 104, 101, 105],
    'customer name': ['Alice Smith', 'bob jones', 'Carol Wilson', 'DAVID BROWN', 'alice smith', 'Eva Davis'],
    'order_value': [150.0, None, 200.0, 180.5, 150.0, None],
    'city': ['New York', 'london', 'Paris', 'NEW YORK', 'New York', 'berlin']
}
df= pd.DataFrame(data)
df.head()
```

```python
df['customer_id'].duplicated().sum()
```

```python
df.drop_duplicates(subset= 'customer_id',inplace= True)
```

```python
df.head()
```

```python
df.reset_index(inplace= True)
```

```python
df.head()
```

```python
df.drop('index',axis=1,inplace= True)
```

```python
df.isnull().sum()
```

```python
df.fillna(df['order_value'].mean(),inplace= True)
df.head()
```

```python
df['city'].str.capitalize()
```

```python
df['customer name'].str.title()
```

```python
df.head()
```

```python
df['customer email']= df['customer name'].str.split().str[0]+'.'+df['customer name'].str.split().str[1]+'@company.com'
```

```python
df.head()
```

### **Task 2: Data Transformation**

```python
def order_category(value):
    if value < 100:
        return 'Low'
    elif value < 200:
        return 'Medium'
    else:
        return 'High'
    
df['order category']= df['order_value'].apply(order_category)
df.head()
```

```python
def country(value):
    if value == 'New York':
        return 'Americas'
    else:
        return 'Europe'
    
df['region']= df['city'].apply(country)
df.head()
```
