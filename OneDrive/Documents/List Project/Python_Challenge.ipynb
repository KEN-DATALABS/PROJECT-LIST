#### **FIRST GAME**

```python
data_blob = """Viktor Reyes | 34 | PHANTOM | active
Elena Cruz | 29 | SHADOW | inactive
Marcus Thorne | 42 | NIGHTFALL | active
Lei Chen | 27 | ECHO | active
Samuel Okonkwo | 31 | REDEMPTION | inactived
Fatima Al-Hassan | 38 | VIPER | active
Dmitri Volkov | 45 | WINTER | inactive
Clara Weiss | 26 | LUMEN | active"""
print(data_blob)
```

```python
suspect_list=data_blob.split('\n')
print(suspect_list)
```

```python
valid_entry=[]
for suspect in suspect_list:
	suspect = suspect.split(' | ')
	print(suspect)
	try:
		suspect[1]= int(suspect[1])
		print(type(suspect[1]))
		new_suspect_list = dict(name=suspect[0],age=suspect[1],codeword=suspect[2],status=suspect[3])
		#print(suspect_list)	
		valid_entry.append(new_suspect_list)
	except ValueError:
		print('Casting failed')
		continue
```

```python
print(valid_entry)
```

```python
def filter_suspects(data, status):
	i = 0
	for suspect in data:
		for value in suspect.values():
			if value == status:
				i+=1
				print(suspect)
	print(f'Total {status} suspects: {i}')
filter_suspects(valid_entry, 'inactive')
```

```python
def report(entries):
    for entry in entries:
        print(f"[{entry['status']}] {entry['name']}(age {entry['age']})-Codeword:{entry['codeword']}")
report(valid_entry)
```
